```python
# fill missing values for trihalomethanes using the mean(since missing is less that 5%) of each potability class

def impute_trihalomethanes(cols):
    trihal = cols[0]
    class_Potability = cols[1]
    
    if pd.isnull(trihal):
        if class_Potability == 1:
            return data[data['Potability']==1].mean()['Trihalomethanes']
        else:
            return data[data['Potability']==0].mean()['Trihalomethanes']
    else:
        return trihal
    
    
    
```

```python
data['Trihalomethanes_mean'] = data[['Trihalomethanes','Potability']].apply(impute_trihalomethanes,axis=1)
```

```python
data['Trihalomethanes'].plot(kind='kde',color='blue')
data['Trihalomethanes_mean'].plot(kind='kde',color='red')
plt.legend(['Trihalomethanes','Trihalomethanes_mean'])
```
