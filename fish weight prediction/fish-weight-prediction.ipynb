## Perfom EDA on Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
## see correlation between the numeric values
```

```python
sns.heatmap(df_fish.corr(),annot=True,cbar=False);
```

from the above visualization we can see that all the numeric values have a strong correlation with the weight

```python
sns.countplot(x=df_fish['Species']);
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df_fish['Weight'],y=df_fish['Length1'])
sns.lineplot(x=df_fish['Weight'],y=df_fish['Length2'])
sns.lineplot(x=df_fish['Weight'],y=df_fish['Length3'])
plt.ylabel('Lengths')
```

from the above we can see that as the fish length goes up so does the weight, also the 3 fish length have close similar values and all follow the close patterns

```python
sns.scatterplot(x=df_fish['Width'],y=df_fish['Height'])
```

as the height increase so does the width increase. we can also see a close pattern that some fish with possible range difference of 2-4 in height can have the same weight.
